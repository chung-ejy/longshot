```python
from transformer.financial_transformer import FinancialTransformer
from database.market import Market
from database.sec import SEC
import pandas as pd
from tqdm import tqdm 
from datetime import datetime
```

```python
sec = SEC()
market = Market()
```

```python
market.connect()
ciks = market.retrieve("sp500")["CIK"]
market.disconnect()
```

```python
sec.connect()
subs = sec.retrieve("subs")
adshs = sec.retrieve_adshs()
sec.disconnect()
```

```python
for cik in ciks:
    print(cik,len(subs[subs["cik"]==int(cik)]))
```

```python
sec.connect()
count = 0
for i in tqdm(range(len(ciks))):
    try:
        cik = int(ciks[i])
        values = subs[(subs["cik"]==cik)]
        all_adshs = values["adsh"].unique()
        for adsh in all_adshs:
            try:
                if adsh not in adshs["adsh"]:
                    num = sec.retrieve_num_data(adsh).drop("_id",axis=1,errors="ignore")
                    values = values.drop("_id",axis=1,errors="ignore")
                    table = FinancialTransformer.transform(values,num)
                    sec.store("filings",table.drop("_id",axis=1,errors="ignore"))
                else:
                    continue
            except Exception as e:
                print(str(e))
                continue
    except Exception as e:
        print(str(e))
        continue
sec.disconnect()
```
